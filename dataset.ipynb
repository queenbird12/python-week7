```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, header=None, names=column_names)

print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("\nClass Distribution:")
print(df['species'].value_counts())


print("\nAverage Sepal Length by Species:")
print(df.groupby('species')['sepal_length'].mean())
plt.figure(figsize=(8, 5))
plt.hist(df['sepal_length'], bins=15, color='lightgreen', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
df.boxplot(column='petal_width', by='species', grid=False)
plt.title('Petal Width by Species')
plt.suptitle('')
plt.ylabel('Petal Width (cm)')
plt.show()

colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}
plt.figure(figsize=(8, 5))
for species, color in colors.items():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], label=species, color=color)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Width by Species')
plt.legend()
plt.grid(True)
plt.show
```
